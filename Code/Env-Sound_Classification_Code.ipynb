```python
# Kaggle-ready notebook code: MFCC + Augmentation + Regularized 2D CNN
# Copy-paste this whole block into one Kaggle notebook cell and run.
# Requirements: librosa, tensorflow, matplotlib, scikit-learn, numpy, soundfile (usually available on Kaggle)

import os
import zipfile
import numpy as np
import librosa
import random
import json
import matplotlib.pyplot as plt
from tqdm import tqdm
from sklearn.model_selection import train_test_split
from tensorflow.keras.utils import to_categorical
import tensorflow as tf
from tensorflow.keras import regularizers
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import (Conv2D, MaxPooling2D, GlobalAveragePooling2D,
                                     Dense, Dropout, BatchNormalization)
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau, ModelCheckpoint

# -------------------------
# 1) Locate dataset on Kaggle (auto-detect common paths)
# -------------------------
def find_data_path():
    # Common locations: /kaggle/input/<dataset>/SoundProject or NatureSounds
    candidates = []
    base = '/kaggle/input'
    if os.path.exists(base):
        for d in os.listdir(base):
            path = os.path.join(base, d)
            if os.path.isdir(path):
                # search for NatureSounds below this
                for root, dirs, files in os.walk(path):
                    if 'NatureSounds' in dirs:
                        return os.path.join(root, 'NatureSounds')
                    # also check for SoundProject zip or folder
                # check if this dataset contains the zip directly
                if any(fname.lower().endswith('soundproject.zip') for fname in os.listdir(path)):
                    # try unzip to /kaggle/working/SoundProject
                    return os.path.join(path, 'SoundProject.zip')
                # sometimes the dataset itself is the extracted folder with NatureSounds
                if 'NatureSounds' in os.listdir(path):
                    return os.path.join(path, 'NatureSounds')
                # else maybe it's directly the files
    # fallback to working directory
    if os.path.exists('SoundProject/NatureSounds'):
        return 'SoundProject/NatureSounds'
    if os.path.exists('/kaggle/working/SoundProject/NatureSounds'):
        return '/kaggle/working/SoundProject/NatureSounds'
    # nothing found
    return None

DATA_LOC = find_data_path()
print("Detected DATA_LOC:", DATA_LOC)

# If dataset is a zip in /kaggle/input, unzip it to /kaggle/working/SoundProject
if DATA_LOC and DATA_LOC.lower().endswith('.zip'):
    zip_path = DATA_LOC
    extract_path = '/kaggle/working/SoundProject'
    print("Unzipping", zip_path, "to", extract_path)
    with zipfile.ZipFile(zip_path, 'r') as zip_ref:
        zip_ref.extractall(extract_path)
    DATA_PATH = os.path.join(extract_path, 'NatureSounds')
else:
    DATA_PATH = DATA_LOC

if DATA_PATH is None or not os.path.exists(DATA_PATH):
    raise FileNotFoundError("Could not find SoundProject/NatureSounds in /kaggle/input or working dir. "
                            "Upload SoundProject.zip or a folder named 'NatureSounds' containing class subfolders.")

print("Using DATA_PATH =", DATA_PATH)
print("Classes:", sorted(os.listdir(DATA_PATH)))

# -------------------------
# 2) Config & helpers
# -------------------------
SAMPLE_RATE = 22050
DURATION = 4
SAMPLES_PER_TRACK = SAMPLE_RATE * DURATION
N_MFCC = 40
AUG_PER_FILE = 2         # how many augmentations to create per original file (set 0 to disable)
SEED = 42
random.seed(SEED)
np.random.seed(SEED)
tf.random.set_seed(SEED)

def load_audio(path, sr=SAMPLE_RATE, duration=DURATION):
    signal, sr = librosa.load(path, sr=sr, duration=duration)
    # If shorter than duration, pad
    if len(signal) < SAMPLES_PER_TRACK:
        signal = np.pad(signal, (0, SAMPLES_PER_TRACK - len(signal)))
    else:
        signal = signal[:SAMPLES_PER_TRACK]
    return signal, sr

# Augmentation functions
def add_noise(signal, noise_factor=0.005):
    noise = np.random.randn(len(signal))
    augmented = signal + noise_factor * noise
    return augmented

def time_stretch(signal, rate=1.0):
    # librosa.effects.time_stretch may change length; handle by warp/resample
    stretched = librosa.effects.time_stretch(signal, rate)
    if len(stretched) < SAMPLES_PER_TRACK:
        stretched = np.pad(stretched, (0, SAMPLES_PER_TRACK - len(stretched)))
    else:
        stretched = stretched[:SAMPLES_PER_TRACK]
    return stretched

def pitch_shift(signal, sr, n_steps):
    shifted = librosa.effects.pitch_shift(signal, sr, n_steps=n_steps)
    if len(shifted) < SAMPLES_PER_TRACK:
        shifted = np.pad(shifted, (0, SAMPLES_PER_TRACK - len(shifted)))
    else:
        shifted = shifted[:SAMPLES_PER_TRACK]
    return shifted

def random_augment(signal, sr):
    # randomly choose an augmentation
    choice = random.choice(['noise', 'stretch', 'pitch'])
    if choice == 'noise':
        return add_noise(signal, noise_factor=random.uniform(0.003, 0.01))
    elif choice == 'stretch':
        rate = random.uniform(0.9, 1.1)
        return time_stretch(signal, rate)
    else:
        steps = random.uniform(-2, 2)
        return pitch_shift(signal, sr, n_steps=steps)

# MFCC extractor (returns transposed [time, n_mfcc])
def extract_mfcc(signal, sr, n_mfcc=N_MFCC):
    mfcc = librosa.feature.mfcc(y=signal, sr=sr, n_mfcc=n_mfcc, n_fft=2048, hop_length=512)
    return mfcc.T  # shape (time_steps, n_mfcc)

# -------------------------
# 3) Build dataset (with augmentation if enabled)
# -------------------------
class_labels = sorted(os.listdir(DATA_PATH))
label_map = {i: label for i, label in enumerate(class_labels)}
print("Label map:", label_map)

X_list = []
y_list = []

print("Extracting MFCCs and creating augmentations...")
for idx, label in enumerate(class_labels):
    folder = os.path.join(DATA_PATH, label)
    files = sorted([f for f in os.listdir(folder) if f.lower().endswith(('.wav', '.mp3'))])
    for fname in tqdm(files, desc=f"Class {label}"):
        fpath = os.path.join(folder, fname)
        try:
            signal, sr = load_audio(fpath)
            # normalize
            if np.max(np.abs(signal)) > 0:
                signal = signal.astype(np.float32) / np.max(np.abs(signal))
            mfcc = extract_mfcc(signal, sr)
            X_list.append(mfcc)
            y_list.append(idx)

            # augmentations
            for a in range(AUG_PER_FILE):
                aug_sig = random_augment(signal, sr)
                if np.max(np.abs(aug_sig)) > 0:
                    aug_sig = aug_sig.astype(np.float32) / np.max(np.abs(aug_sig))
                mfcc_aug = extract_mfcc(aug_sig, sr)
                X_list.append(mfcc_aug)
                y_list.append(idx)

        except Exception as e:
            print("Error:", fpath, e)

X = np.array(X_list, dtype=np.float32)  # shape (N, time, n_mfcc)
y = np.array(y_list, dtype=np.int32)
print("Raw MFCC array shape:", X.shape, "labels shape:", y.shape)

# Determine max frames and pad/truncate to fixed size (usually ~173)
max_frames = max([x.shape[0] for x in X])
# We will set target_frames to that max (usually 173)
TARGET_FRAMES = max_frames
print("TARGET_FRAMES:", TARGET_FRAMES)

# Pad/truncate each MFCC to TARGET_FRAMES
X_padded = []
for mf in X:
    if mf.shape[0] < TARGET_FRAMES:
        pad_width = TARGET_FRAMES - mf.shape[0]
        mf_p = np.pad(mf, ((0,pad_width),(0,0)), mode='constant')
    else:
        mf_p = mf[:TARGET_FRAMES, :]
    X_padded.append(mf_p)

X = np.array(X_padded, dtype=np.float32)
# reshape for Conv2D: (samples, frames, n_mfcc, 1)
X = X[..., np.newaxis]
y_cat = to_categorical(y, num_classes=len(class_labels))

print("Final X shape (for CNN):", X.shape)
print("Final y shape:", y_cat.shape)

# Shuffle dataset
perm = np.random.permutation(len(X))
X = X[perm]
y_cat = y_cat[perm]
y = y[perm]

# -------------------------
# 4) Train/test split
# -------------------------
TEST_SIZE = 0.2
X_train, X_test, y_train, y_test = train_test_split(X, y_cat, test_size=TEST_SIZE, random_state=SEED, stratify=y)
print("Train shape:", X_train.shape, "Test shape:", X_test.shape)

# -------------------------
# 5) Build regularized CNN model
# -------------------------
input_shape = X_train.shape[1:]  # (frames, n_mfcc, 1)
print("Input shape:", input_shape)

weight_decay = 1e-4
model = Sequential([
    Conv2D(32, (3,3), activation='relu', padding='same', kernel_regularizer=regularizers.l2(weight_decay), input_shape=input_shape),
    BatchNormalization(),
    MaxPooling2D((2,2)),
    Dropout(0.2),

    Conv2D(64, (3,3), activation='relu', padding='same', kernel_regularizer=regularizers.l2(weight_decay)),
    BatchNormalization(),
    MaxPooling2D((2,2)),
    Dropout(0.25),

    Conv2D(128, (3,3), activation='relu', padding='same', kernel_regularizer=regularizers.l2(weight_decay)),
    BatchNormalization(),
    MaxPooling2D((2,2)),
    Dropout(0.3),

    GlobalAveragePooling2D(),
    Dense(128, activation='relu', kernel_regularizer=regularizers.l2(weight_decay)),
    Dropout(0.4),
    Dense(len(class_labels), activation='softmax')
])

model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=1e-3),
              loss='categorical_crossentropy',
              metrics=['accuracy'])
model.summary()

# -------------------------
# 6) Callbacks
# -------------------------
callbacks = [
    EarlyStopping(monitor='val_loss', patience=8, restore_best_weights=True, verbose=1),
    ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=4, min_lr=1e-6, verbose=1),
    ModelCheckpoint('/kaggle/working/best_sound_model.h5', monitor='val_loss', save_best_only=True, verbose=1)
]

# -------------------------
# 7) Train
# -------------------------
EPOCHS = 60
BATCH_SIZE = 8

history = model.fit(
    X_train, y_train,
    validation_data=(X_test, y_test),
    epochs=EPOCHS,
    batch_size=BATCH_SIZE,
    callbacks=callbacks,
    verbose=2
)

# -------------------------
# 8) Evaluate & Save
# -------------------------
loss, acc = model.evaluate(X_test, y_test, verbose=0)
print(f"\nFinal Test Accuracy: {acc*100:.2f}%")
print("Final Test Loss:", loss)

# Save final model and label_map
model.save('/kaggle/working/nature_sound_mfcc_model_kaggle.h5')
with open('/kaggle/working/label_map.json', 'w') as f:
    json.dump(label_map, f)

# -------------------------
# 9) Plot training curves
# -------------------------
plt.figure(figsize=(10,4))
plt.plot(history.history['accuracy'], label='train_acc')
plt.plot(history.history['val_accuracy'], label='val_acc')
plt.title('Accuracy')
plt.legend()
plt.grid(True)
plt.show()

plt.figure(figsize=(10,4))
plt.plot(history.history['loss'], label='train_loss')
plt.plot(history.history['val_loss'], label='val_loss')
plt.title('Loss')
plt.legend()
plt.grid(True)
plt.show()

# -------------------------
# 10) Quick inference demo (top-3 predictions) on test set
# -------------------------
import numpy as np
from tensorflow.keras.models import load_model

m = load_model('/kaggle/working/best_sound_model.h5')
# pick a random test sample
idx = np.random.randint(0, X_test.shape[0])
sample = X_test[idx:idx+1]
probs = m.predict(sample)[0]
top3 = probs.argsort()[-3:][::-1]
print("Top-3 predictions:")
for i in top3:
    print(f"  {label_map[i]}: {probs[i]*100:.2f}%")

# Save the training history to disk for later reporting
import pickle
with open('/kaggle/working/history.pkl', 'wb') as f:
    pickle.dump(history.history, f)

print("\nSaved model and label map to /kaggle/working/. Download from Kaggle UI.")

# ------------------------
```
